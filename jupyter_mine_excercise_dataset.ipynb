```python
import csv, sqlite3

con = sqlite3.connect("data.db")
cur = con.cursor()
```

```python
%sql sqlite:///data.db
```

```python
import pandas
df = pandas.read_csv('data.csv')
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")
```

```python
%sql select * from chicago_socioeconomic_data limit 5
```

```python
%sql select * from chicago_socioeconomic_data where [HARDSHIP INDEX] > 50
```

```python
%sql select [COMMUNITY AREA NAME] from chicago_socioeconomic_data where [hardship index] = (select max([hardship index]) from chicago_socioeconomic_data )
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

income_vs_hardship=%sql select [per capita income ] as pi, [hardship index] as hi FROM chicago_socioeconomic_data;
plot=sns.jointplot(x='pi',y='hi',data=income_vs_hardship.DataFrame())
```
